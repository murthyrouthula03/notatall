```python
import pandas as pd
import numpy as np
import os,sys
import re
pwd = os.getcwd()
```

```python
# reading dataset
kb_df = pd.read_csv(pwd+"//kabitakitchen.csv",encoding_errors='ignore')
kb_df
```

### Cleaning data

```python
# replacing enter with ';'
kb_df['commentText'] = kb_df['commentText'].str.replace('\n',';')
kb_df
```

```python
# kb_df['only_words'] = kb_df['commentText'].apply(lambda x: re.findall(r'\w+',x))
# kb_df
```

```python
# getting stop words
with open(pwd+'//hinglish_stopwords_filter.txt','r') as sw:
    stw_list = sw.readlines()
stw_l = list(map(lambda x: x.strip(),stw_list))
```

# Vectorizations

### TFIDF Vectorizer

```python
from sklearn.feature_extraction.text import TfidfVectorizer
tvectorizer = TfidfVectorizer(lowercase=True,stop_words=stw_l,max_features = 500)
Z = tvectorizer.fit_transform(kb_df['commentText'])
word_vec_arr = Z.toarray()
word_vec_cols = tvectorizer.get_feature_names()
```

```python
tfidf_500_df = pd.DataFrame(word_vec_arr,columns=word_vec_cols)
tfidf_500_df
```

```python
#tfidf_500_df.to_csv(pwd+'//tfidf_500_vectors.csv',index=False)
```

### Count Vectorizer

```python
from sklearn.feature_extraction.text import CountVectorizer
cvectorizer = CountVectorizer(lowercase=True,stop_words=stw_l,max_features = 500)
X = cvectorizer.fit_transform(kb_df['commentText'])
cword_vec_cols = cvectorizer.get_feature_names()
cword_vec_arr = X.toarray()
```

```python
cv_500_df = pd.DataFrame(cword_vec_arr,columns=cword_vec_cols)
cv_500_df
```

```python
#cv_500_df.to_csv(pwd+'//cv_500_vectors.csv',index=False)
```

### Term Frequency Vectorizer

```python
from sklearn.feature_extraction.text import TfidfVectorizer
tfvectorizer = TfidfVectorizer(lowercase=True,stop_words=stw_l,max_features = 500,use_idf=False)
tf = tfvectorizer.fit_transform(kb_df['commentText'])
tfword_vec_arr = tf.toarray()
tfword_vec_cols = tfvectorizer.get_feature_names()
```

```python
tf_500_df = pd.DataFrame(tfword_vec_arr,columns=tfword_vec_cols)
tf_500_df
```

```python
#tf_500_df.to_csv(pwd+'//tf_500_vectors.csv',index=False)
```

# Modelling

```python
# importing algorithms
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LogisticRegression
from sklearn.neighbors import KNeighborsClassifier
```

### data split for TFIDF data

```python
x_train,x_test,y_train,y_test = train_test_split(tfidf_500_df,kb_df['Labels'],test_size=0.30,
                                                 random_state=21,stratify=kb_df['Labels'])
```

### Logistic regression for TFIDF data

```python
# applying logistic regression
tv_lr_model = LogisticRegression()
tv_lr_model.fit(x_train,y_train)
tv_lr_pred_val = tv_lr_model.predict(x_test)
tv_lr_model.score(x_test,y_test)
```

```python
tv_lr_dict = {}
for x in tv_lr_pred_val:
    try:
        tv_lr_dict[str(x)] += 1
    except:
        tv_lr_dict[str(x)] = 0
        tv_lr_dict[str(x)] += 1
        
tv_lr_dict
```

```python
tv_lr_real_dict = {}
for x in y_test:
    try:
        tv_lr_real_dict[str(x)] += 1
    except:
        tv_lr_real_dict[str(x)] = 0
        tv_lr_real_dict[str(x)] += 1
        
tv_lr_real_dict
```

### KNN for TFIDF data

```python
# applying KNN algorithm
# taking squareroot(n) as k value. as total rows are 4900, k-value = 70
tv_knn_model = KNeighborsClassifier(n_neighbors=70)
tv_knn_model.fit(x_train,y_train)
tv_knn_pred_val = tv_knn_model.predict(x_test)
tv_knn_model.score(x_test,y_test)
```

```python
# checking the accuracies of neighbors in [3, 4, 5, 6, 7, 8] for KNN
neighbors_list = [3, 4, 5, 6, 7, 8]
for x in neighbors_list:
    tv_knn_model = KNeighborsClassifier(n_neighbors=x)
    tv_knn_model.fit(x_train,y_train)
    print("Accuracy of KNN model for {} neighbors is:".format(str(x)), 
          tv_knn_model.score(x_test,y_test))
```
