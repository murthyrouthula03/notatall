```python
import pandas as pd
import numpy as np
import os,sys
import re
pwd = os.getcwd()
```

```python
# reading dataset
kb_df = pd.read_csv(pwd+"//kabitakitchen.csv")
kb_df
```

### Cleaning data

```python
# replacing enter with ';'
kb_df['commentText'] = kb_df['commentText'].str.replace('\n',';')
kb_df
```

```python
# kb_df['only_words'] = kb_df['commentText'].apply(lambda x: re.findall(r'\w+',x))
# kb_df
```

### TFIDF Vectorization

```python
with open(pwd+'//hinglish_stopwords_filter.txt','r') as sw:
    stw_list = sw.readlines()
stw_l = list(map(lambda x: x.strip(),stw_list))
```

```python
from sklearn.feature_extraction.text import TfidfVectorizer
vectorizer1 = TfidfVectorizer(lowercase=True,stop_words=stw_l,max_features = 40)
X = vectorizer1.fit_transform(kb_df['commentText'])
print(X.toarray())
print(vectorizer1.get_feature_names())
```

```python
kb_arr = X.toarray()
for y in range(0,5):
    for x in kb_arr[y]:
        if x > 0:
            print(x)
    print(len(kb_arr[y]))
    print("***********************")
```

```python
from sklearn.feature_extraction.text import TfidfVectorizer
vectorizer2 = TfidfVectorizer(lowercase=True,stop_words=stw_l,max_features = 100)
Y = vectorizer2.fit_transform(kb_df['commentText'])
print(Y.toarray())
print(vectorizer2.get_feature_names())
```

```python
kb_arry = Y.toarray()
for y in range(0,5):
    for x in kb_arry[y]:
        if x > 0:
            print(x)
    print(len(kb_arry[y]))
    print("***********************")
```

```python
from sklearn.feature_extraction.text import TfidfVectorizer
vectorizer3 = TfidfVectorizer(lowercase=True,stop_words=stw_l,max_features = 500)
Z = vectorizer3.fit_transform(kb_df['commentText'])
word_vec_arr = Z.toarray()
word_vec_cols = vectorizer3.get_feature_names()
```

```python
tfidf_500_df = pd.DataFrame(word_vec_arr,columns=word_vec_cols)
tfidf_500_df
```

```python
tfidf_500_df.to_csv(pwd+'//tfidf_500_vectors.csv',index=False)
```

```python
kb_arrz = Z.toarray()
for y in range(0,5):
    for x in kb_arrz[y]:
        if x > 0:
            print(x)
    print(len(kb_arrz[y]))
    print("***********************")
```

```python
from sklearn.feature_extraction.text import CountVectorizer
cvectorizer = CountVectorizer(lowercase=True,stop_words=stw_l,max_features = 500)
X = cvectorizer.fit_transform(kb_df['commentText'])
cword_vec_cols = cvectorizer.get_feature_names()
cword_vec_arr = X.toarray()
```

```python
cv_500_df = pd.DataFrame(cword_vec_arr,columns=cword_vec_cols)
cv_500_df
```

```python
cv_500_df.to_csv(pwd+'//cv_500_vectors.csv',index=False)
```
