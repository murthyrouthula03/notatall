```python
try:
    import pandas as pd
    import numpy as np
    import os,sys
    import re
    # importing algorithms
    from sklearn.model_selection import train_test_split
    from sklearn.linear_model import LogisticRegression
    from sklearn.neighbors import KNeighborsClassifier
    from sklearn.naive_bayes import GaussianNB
    from sklearn.naive_bayes import MultinomialNB
    from sklearn.naive_bayes import BernoulliNB
    from sklearn import svm
    from sklearn.tree import DecisionTreeClassifier
    from sklearn.ensemble import RandomForestClassifier
except Exception as e:
    print("Error is due to",e)
pwd = os.getcwd()
```

```python
# Reading labels
labels_df = pd.read_csv(pwd+"//Datasets//Kabita//Input//kabita_dataset_labels.csv")
labels_df
```

### Data split for TFIDF data

```python
# reading dataset
tfidf_500_df = pd.read_csv(pwd+"//Datasets//Kabita//BagOfWords//tfidf_500_vectors.csv",encoding_errors='ignore')
tfidf_500_df
```

```python
x_train,x_test,y_train,y_test = train_test_split(tfidf_500_df,labels_df['kabita_labels'],test_size=0.30,
                                                 random_state=21,stratify=labels_df['kabita_labels'])
```

```python
# Logistic regression for TFIDF data
tv_lr_model = LogisticRegression()
tv_lr_model.fit(x_train,y_train)
tv_lr_pred_val = tv_lr_model.predict(x_test)
print("Accuracy of Logistic Regression for TFIDF is:", tv_lr_model.score(x_test,y_test))

# KNN for TFIDF data
# taking squareroot(n) as k value. as total rows are 4900, k-value = 70
tv_knn_model = KNeighborsClassifier(n_neighbors=70)
tv_knn_model.fit(x_train,y_train)
tv_knn_pred_val = tv_knn_model.predict(x_test)
print("Accuracy of KNN for TFIDF for 70 neghbors is:", tv_knn_model.score(x_test,y_test))
# checking the accuracies of neighbors in [3, 4, 5, 6, 7, 8] for KNN
neighbors_list = [3, 4, 5, 6, 7, 8]
for x in neighbors_list:
    tv_knn_model = KNeighborsClassifier(n_neighbors=x)
    tv_knn_model.fit(x_train,y_train)
    print("Accuracy of KNN model for {} neighbors is:".format(str(x)), 
          tv_knn_model.score(x_test,y_test))
```
