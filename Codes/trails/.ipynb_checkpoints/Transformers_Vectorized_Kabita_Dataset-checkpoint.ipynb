```python
import pandas as pd
import numpy as np
import os,sys
import re
pwd = os.getcwd()
```

```python
# reading dataset
kb_df = pd.read_csv(pwd+"//kabitakitchen.csv", encoding_errors='ignore')
kb_df
```

```python
kb_df.shape
```

```python
def preprocess_text(sen):
    # Removing html tags
    sentence = remove_tags(sen)

    # Remove punctuations and numbers
    sentence = re.sub('[^a-zA-Z]', ' ', sentence)

    # Single character removal
    sentence = re.sub(r"\s+[a-zA-Z]\s+", ' ', sentence)

    # Removing multiple spaces
    sentence = re.sub(r'\s+', ' ', sentence)

    return sentence
```

```python
TAG_RE = re.compile(r'<[^>]+>')

def remove_tags(text):
    return TAG_RE.sub('', text)
```

```python
comments = []
sentences = list(kb_df['commentText'])
for sen in sentences:
    comments.append(preprocess_text(sen))
```

```python
comments
```

### getting word embedding using XLM transformer model

```python
from sentence_transformers import SentenceTransformer
#sentences = ["This is an example sentence", "Each sentence is converted"]

model = SentenceTransformer('sentence-transformers/stsb-xlm-r-multilingual')
embeddings = model.encode(comments)
print(embeddings)
```

```python
xlm_df = pd.DataFrame(embeddings)
xlm_df
```

```python
xlm_df.to_csv(pwd+"//xlm_vectorized_kabita_dataset.csv",index=False)
```

```python
from sentence_transformers import SentenceTransformer
#sentences = ["This is an example sentence", "Each sentence is converted"]

model = SentenceTransformer('sentence-transformers/bert-base-nli-mean-tokens')
embeddings = model.encode(comments)
print(embeddings)
```
