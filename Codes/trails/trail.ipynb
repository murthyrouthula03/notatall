```python
import pandas as pd
import numpy as np
import os,sys
import re
pwd = os.getcwd()
```

```python
# reading dataset
kb_df1 = pd.read_csv(pwd+"//Datasets//Kabita//Input//kabitakitchen.csv", encoding_errors='ignore')
kb_df1
```

```python
# taking first 4000 rows
kb_df = kb_df1.head(4000)
kb_df
```

```python
# taking last 900 rows for final test
kb_df_final = kb_df1.tail(900)
kb_df_final
```

```python
kb_df['Labels'] = kb_df['Labels']-1
kb_df_final['Labels'] = kb_df_final['Labels']-1
```

```python
X=list(kb_df['commentText'])
y=list(kb_df['Labels'])
X_final = list(kb_df_final['commentText'])
y_final=list(kb_df_final['Labels'])
```

```python
print(len(X),len(y),len(X_final),len(y_final))
```

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size = 0.20, random_state = 30)
```

```python
from transformers import BertTokenizer
tokenizer = BertTokenizer.from_pretrained('bert-base-uncased')
```

```python
train_encodings = tokenizer(X_train, truncation=True, padding=True)
test_encodings = tokenizer(X_test, truncation=True, padding=True)
test_encodings_final = tokenizer(X_final, truncation=True, padding=True)
```

```python
print(test_encodings)
```

```python
import tensorflow as tf

train_dataset = tf.data.Dataset.from_tensor_slices((
    dict(train_encodings),
    y_train
))

test_dataset = tf.data.Dataset.from_tensor_slices((
    dict(test_encodings),
    y_test
))

test_dataset_final = tf.data.Dataset.from_tensor_slices((
    dict(test_encodings_final),
    y_final
))
```

```python
print(test_dataset)
```

```python
from transformers import TFBertForSequenceClassification, TFTrainer, TFTrainingArguments

training_args = TFTrainingArguments(
    output_dir='./Berttrailresults',          # output directory
    num_train_epochs=2,              # total number of training epochs
    per_device_train_batch_size=8,  # batch size per device during training
    per_device_eval_batch_size=16,   # batch size for evaluation
    warmup_steps=500,                # number of warmup steps for learning rate scheduler
    weight_decay=0.01,               # strength of weight decay
    logging_dir='./Berttraillogs',           # directory for storing logs
    logging_steps=10,
    eval_steps=10
)
```

```python
with training_args.strategy.scope():
    model = TFBertForSequenceClassification.from_pretrained("bert-base-uncased", num_labels=7)

trainer = TFTrainer(
    model=model,                         # the instantiated ðŸ¤— Transformers model to be trained
    args=training_args,                  # training arguments, defined above
    train_dataset=train_dataset,         # training dataset
    eval_dataset=test_dataset             # evaluation dataset
)

trainer.train()
```

```python
trainer.evaluate(test_dataset)
```

```python
output=trainer.predict(test_dataset)
```
