```python
# importing Libraries
try:
    import os, sys
    import pandas as pd
    import numpy as np
    from sklearn.preprocessing import StandardScaler
    from sklearn.decomposition import PCA
    import matplotlib.pyplot as plt
    import seaborn as sns
except Exception as e:
    print("Error because of",e)
pwd = os.getcwd()
```

```python
def scree_graph(df, data_set):
    scaler_model=StandardScaler()
    scaled_data= scaler_model.fit_transform(df)
    #print(scaled_data)
    # Doing  PCA giving number of Components(dimensions)
    pca_comp=PCA(n_components=20)
    x_pca_comp=pca_comp.fit_transform(scaled_data)
    variances_list = pca_comp.explained_variance_ratio_
    plt.plot(list(range(1,21)), variances_list, 'o-', color='black')
    plt.title('Scree Plot of '+data_set+" data")
    plt.xticks(range(21))
    plt.xlabel('Principal Component')
    plt.ylabel('Variance Explained')
    plt.show()
```

### Bag Of Words Scree Plots

```python
# TFIDF vectorized data
df = pd.read_csv(pwd+"//Datasets//Nisha//BagOfWords//tfidf_500_vectors.csv")
scree_graph(df, 'TFIDF')
```

```python
# Count Vectorizer data
df = pd.read_csv(pwd+"//Datasets//Nisha//BagOfWords//cv_500_vectors.csv")
scree_graph(df, 'Count Vectorized data')
```

```python
# Term Frequency Vectorizer data
df = pd.read_csv(pwd+"//Datasets//Nisha//BagOfWords//tf_500_vectors.csv")
scree_graph(df, 'term Frequency Vectorized data')
```
