```python
try:
    import os, sys
    import pandas as pd
    import seaborn as sns
    import matplotlib.pyplot as plt
    import numpy as np
except Exception as e:
    print("Error due to", e)
```

```python
pwd = os.getcwd()
# Reading preprocessing file
kb_df = pd.read_csv(pwd+"//Datasets//Kabita//Input//kabita_preprocessing.csv",encoding_errors='ignore')
kb_df
```

```python
# info of data
kb_df.info()
```

commentText is Nominal categorical data as there is no natural order and are just comments.

Labels, stopwords, upper, hashtags, word_count, char_count, numerics are the info of number of their presence in comments. It is Discrete Numerical data.

avg_word is (char_count - spaces)/word_count. It is Continuous Numerical data.

### Data Visualization

```python
# univariate plot of labels
sns.countplot(x = 'Labels', data=kb_df)
plt.show()
```

```python
# univariate plot of stopwords
sns.countplot(x = 'stopwords', data=kb_df)
plt.show()
```

```python
# univariate plot of upper words
sns.countplot(x = 'upper', data=kb_df)
plt.show()
```

```python
# univariate plot of hashtags
sns.countplot(x = 'hastags', data=kb_df)
plt.show()
```

```python
# univariate plot of word
sns.boxplot(x = 'word_count', data=kb_df)
plt.show()
```

```python
# univariate plot of word
sns.boxplot(x = 'char_count', data=kb_df)
plt.show()
```

```python
# univariate plot of avg words
sns.boxplot(x = 'avg_word', data=kb_df)
plt.show()
```

```python
#relation between Labels and stopwords.
sns.boxplot(x = 'Labels', y = 'stopwords', data = kb_df)
plt.xlabel("Labels")
plt.ylabel("Stopwords")
plt.show()
```

```python
#relation between Labels and upperwords.
sns.boxplot(x = 'Labels', y = 'upper', data = kb_df)
plt.xlabel("Labels")
plt.ylabel("Upper Words")
plt.show()
```

```python
#relation between Labels and word count.
sns.boxplot(x = 'Labels', y = 'word_count', data = kb_df)
plt.xlabel("Labels")
plt.ylabel("Word Count")
plt.show()
```

```python
#relation between Labels and Char count.
sns.boxplot(x = 'Labels', y = 'char_count', data = kb_df)
plt.xlabel("Labels")
plt.ylabel("Character Count")
plt.show()
```

```python
#relation between Labels and Average words.
sns.boxplot(x = 'Labels', y = 'avg_word', data = kb_df)
plt.xlabel("Labels")
plt.ylabel("Average Word")
plt.show()
```

```python
# lm plot - Multivariate plot of word count, char count, labels
sns.lmplot(x='word_count',y = 'char_count',hue='Labels',data=kb_df,fit_reg=False)
plt.show()
```

```python
# lm plot - Multivariate plot of word count, Average word, labels
sns.lmplot(x='word_count',y = 'avg_word',hue='Labels',data=kb_df,fit_reg=False)
plt.show()
```

```python
# lm plot - Multivariate plot of char count, Average word, labels
sns.lmplot(x='char_count',y = 'avg_word',hue='Labels',data=kb_df,fit_reg=False)
plt.show()
```

```python
# lm plot - Multivariate plot of word count, stopwords, labels
sns.lmplot(x='word_count',y = 'stopwords',hue='Labels',data=kb_df,fit_reg=False)
plt.show()
```
