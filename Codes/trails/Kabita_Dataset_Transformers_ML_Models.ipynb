```python
try:
    import pandas as pd
    import numpy as np
    from numpy import set_printoptions
    import os,sys
    import re
    # importing algorithms
    from sklearn.model_selection import train_test_split
    from sklearn.linear_model import LogisticRegression
    from sklearn.neighbors import KNeighborsClassifier
    from sklearn.naive_bayes import GaussianNB
    from sklearn.naive_bayes import MultinomialNB
    from sklearn.naive_bayes import BernoulliNB
    from sklearn import svm
    from sklearn.tree import DecisionTreeClassifier
    from sklearn.ensemble import RandomForestClassifier
    from sklearn.metrics import confusion_matrix, classification_report
    from sklearn.preprocessing import MinMaxScaler
except Exception as e:
    print("Error is due to",e)
pwd = os.getcwd()
```

```python
# Reading labels
labels_df = pd.read_csv(pwd+"//Datasets//Kabita//Input//kabita_dataset_labels.csv")
labels_df
```

### Modelling on Transformers vectorized data

### BERT Sentence transformer Base Model

```python
# reading dataset
bert_df = pd.read_csv(pwd+"//Datasets//Kabita//SentenceTransformers//bert_vectorized_kabita_dataset.csv")
x_train,x_test,y_train,y_test = train_test_split(bert_df,labels_df['kabita_labels'],test_size=0.30,
                                                 random_state=12,stratify=labels_df['kabita_labels'])
print(70*"=")
# KNN
# checking the accuracies of neighbors in [3, 4, 5, 6, 7, 8] for KNN
neighbors_list = [3, 4, 5, 6, 7, 8]
for x in neighbors_list:
    bert_knn_model = KNeighborsClassifier(n_neighbors=x)
    bert_knn_model.fit(x_train,y_train)
    print("Accuracy of KNN model for {} neighbors is:".format(str(x)), 
          bert_knn_model.score(x_test,y_test))
    pred = bert_knn_model.predict(x_test)
    print("Confusion Matrix of KNN Model for {} neighbors is:\n".format(str(x)),confusion_matrix(y_test,pred))
    print("Classification Report of KNN Model for {} neighbors is:\n".format(str(x)),classification_report(y_test,pred))
    print(70*"=")

#sys.exit()
# applying KNN algorithm
# taking squareroot(n) as k value. as total rows are 4900, k-value = 70
bert_knn_model = KNeighborsClassifier(n_neighbors=70)
bert_knn_model.fit(x_train,y_train)
print("Accuracy of KNN model for 70 neighbors is:", bert_knn_model.score(x_test,y_test))
pred = bert_knn_model.predict(x_test)
print("Confusion Matrix of KNN Model for 70 neighbors is:\n".format(str(x)),confusion_matrix(y_test,pred))
print("Classification Report of KNN Model for 70 neighbors is:\n".format(str(x)),classification_report(y_test,pred))
print(70*"=")

# Logistic Regression
bert_lr_model = LogisticRegression(max_iter=2000)
bert_lr_model.fit(x_train,y_train)
bert_lr_pred_val = bert_lr_model.predict(x_test)
print("Accuracy of Logistic Regression is:", bert_lr_model.score(x_test,y_test))
print("Confusion Matrix of Logistic Regression Model is:\n",confusion_matrix(y_test,bert_lr_pred_val))
print("Classification Report of Logistic Regression Model is:\n",classification_report(y_test,bert_lr_pred_val))
print(70*"=")

# Gaussian Naive Bayes
bert_gnb_model = GaussianNB()
bert_gnb_model.fit(x_train,y_train)
bert_gnb_pred_val = bert_gnb_model.predict(x_test)
print("Accuracy of Gaussian Naive Bayes is:", bert_gnb_model.score(x_test,y_test))
print("Confusion Matrix of Gaussian Naive Bayes Model is:\n",confusion_matrix(y_test,bert_gnb_pred_val))
print("Classification Report of Gaussian Naive Bayes Model is:\n",classification_report(y_test,bert_gnb_pred_val))
print(70*"=")

# Bernoulli Naive Bayes
bert_bnb_model = BernoulliNB()
bert_bnb_model.fit(x_train,y_train)
bert_bnb_pred_val = bert_bnb_model.predict(x_test)
print("Accuracy of Bernoulli Naive Bayes is:", bert_bnb_model.score(x_test,y_test))
print("Confusion Matrix of Bernoulli Naive Bayes Model is:\n",confusion_matrix(y_test,bert_bnb_pred_val))
print("Classification Report of Bernoulli Naive Bayes Model is:\n",classification_report(y_test,bert_bnb_pred_val))
print(70*"=")

# Support Vector Machine Classifier
svm_kernels = ['linear', 'poly', 'rbf', 'sigmoid']
for a_kernel in svm_kernels:
    bert_svm_model = svm.SVC(kernel=a_kernel)
    bert_svm_model.fit(x_train,y_train)
    bert_svm_pred_val = bert_svm_model.predict(x_test)
    print("Working on SVM Kernal:", a_kernel)
    print("Accuracy of SVM model is:", bert_svm_model.score(x_test,y_test))
    print("Confusion Matrix of SVM Model is:\n",confusion_matrix(y_test,bert_svm_pred_val))
    print("Classification Report of SVM Model is:\n",classification_report(y_test,bert_svm_pred_val))
    print(70*"=")

# Decision Tree Classifier
for x in range(1,21):
    bert_dt_model = DecisionTreeClassifier(random_state=3, max_depth=x)
    bert_dt_model.fit(x_train,y_train)
    bert_dt_pred_val = bert_dt_model.predict(x_test)
    print("Accuracy of Decision Tree with max_depth - {} is:".format(str(x)), bert_dt_model.score(x_test,y_test))
    print("Confusion Matrix of Decision Tree Model with max_depth - {} is:\n".format(str(x)),confusion_matrix(y_test,bert_dt_pred_val))
    print("Classification Report of Decision Tree Model with max_depth - {} is:\n".format(str(x)),classification_report(y_test,bert_dt_pred_val))
    print(70*"=")

    
# Random Forest
for x in range(1,21):
    bert_rf_model = RandomForestClassifier(max_depth=x, random_state=3)
    bert_rf_model.fit(x_train,y_train)
    bert_rf_pred_val = bert_rf_model.predict(x_test)
    print("Accuracy of Random Forest with max_depth - {} is:".format(str(x)), bert_rf_model.score(x_test,y_test))
    print("Confusion Matrix of Random Forest Model with max_depth - {} is:\n".format(str(x)),confusion_matrix(y_test,bert_rf_pred_val))
    print("Classification Report of Random Forest Model with max_depth - {} is:\n".format(str(x)),classification_report(y_test,bert_rf_pred_val))
    print(70*"=")
    
# scaling using MinMax scaler
mms_scale=MinMaxScaler(feature_range=(0,10))
m_train=mms_scale.fit_transform(x_train)
m_test=mms_scale.fit_transform(x_test)
set_printoptions(precision=3)
# Multinomial Naive Bayes
bert_mnb_model = MultinomialNB()
bert_mnb_model.fit(m_train,y_train)
bert_mnb_pred_val = bert_mnb_model.predict(m_test)
print("Accuracy of Multinomial Naive Bayes is:", bert_mnb_model.score(m_test,y_test))
print("Confusion Matrix of Multinomial Naive Bayes Model is:\n",confusion_matrix(y_test,bert_mnb_pred_val))
print("Classification Report of Multinomial Naive Bayes Model is:\n",classification_report(y_test,bert_mnb_pred_val))
print(70*"=")
```
