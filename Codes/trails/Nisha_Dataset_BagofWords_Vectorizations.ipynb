```python
import pandas as pd
import numpy as np
import os,sys
import re
pwd = os.getcwd()
```

```python
# reading dataset
kb_df = pd.read_csv(pwd+"//Datasets//Nisha//Input//Nishafin.csv",encoding_errors='ignore')
kb_df
```

```python
new_labels = kb_df[['Labels']]
new_labels.to_csv(pwd+'//Datasets//Nisha//Input//Nisha_dataset_labels.csv',index=False)
```

### Cleaning data

```python
# replacing enter with ';'
kb_df['commentText'] = kb_df['commentText'].str.replace('\n',';')
kb_df
```

```python
# removing emojis
import emoji

def remove_emoji(text):
    return emoji.get_emoji_regexp().sub(u'', text)
```

```python
# remove special characters 
import re

def remove_spechar(text):
    r = re.sub(r'\W+', ' ', text)
    return r.strip()
```

```python
X = list(map(remove_emoji,list(kb_df['commentText'])))
X = list(map(remove_spechar,X))
```

```python
new_kb_df = pd.DataFrame(X,columns=['commentText'])
new_kb_df
```

```python
# getting stop words
with open(pwd+'//Datasets//Nisha//Input//hinglish_stopwords_filter.txt','r') as sw:
    stw_list = sw.readlines()
stw_l = list(map(lambda x: x.strip(),stw_list))
```

# Vectorizations

### TFIDF Vectorizer

```python
from sklearn.feature_extraction.text import TfidfVectorizer
tvectorizer = TfidfVectorizer(lowercase=True,stop_words=stw_l,max_features = 500)
Z = tvectorizer.fit_transform(new_kb_df['commentText'])
word_vec_arr = Z.toarray()
word_vec_cols = tvectorizer.get_feature_names()
```

```python
tfidf_500_df = pd.DataFrame(word_vec_arr,columns=word_vec_cols)
tfidf_500_df
```

```python
tfidf_500_df.to_csv(pwd+'//Datasets//Nisha//BagOfWords//tfidf_500_vectors.csv',index=False)
```

### Count Vectorizer

```python
from sklearn.feature_extraction.text import CountVectorizer
cvectorizer = CountVectorizer(lowercase=True,stop_words=stw_l,max_features = 500)
X = cvectorizer.fit_transform(kb_df['commentText'])
cword_vec_cols = cvectorizer.get_feature_names()
cword_vec_arr = X.toarray()
```

```python
cv_500_df = pd.DataFrame(cword_vec_arr,columns=cword_vec_cols)
cv_500_df
```

```python
cv_500_df.to_csv(pwd+'//Datasets//Nisha//BagOfWords//cv_500_vectors.csv',index=False)
```

### Term Frequency Vectorizer

```python
from sklearn.feature_extraction.text import TfidfVectorizer
tfvectorizer = TfidfVectorizer(lowercase=True,stop_words=stw_l,max_features = 500,use_idf=False)
tf = tfvectorizer.fit_transform(kb_df['commentText'])
tfword_vec_arr = tf.toarray()
tfword_vec_cols = tfvectorizer.get_feature_names()
```

```python
tf_500_df = pd.DataFrame(tfword_vec_arr,columns=tfword_vec_cols)
tf_500_df
```

```python
tf_500_df.to_csv(pwd+'//Datasets//Nisha//BagOfWords//tf_500_vectors.csv',index=False)
```
