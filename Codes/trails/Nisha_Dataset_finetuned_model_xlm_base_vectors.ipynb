```python
import pandas as pd
import numpy as np
import os,sys
import re
pwd = os.getcwd()
```

```python
# reading dataset
kb_df = pd.read_csv(pwd+"//Datasets//Nisha//Input//Nishafin.csv", encoding_errors='ignore')
kb_df
```

```python
kb_df.shape
```

```python
# replacing enter with ';'
kb_df['commentText'] = kb_df['commentText'].str.replace('\n',';')
kb_df
```

```python
# removing emojis
import emoji

def remove_emoji(text):
    return emoji.get_emoji_regexp().sub(u'', text)
```

```python
# remove special characters 
import re

def remove_spechar(text):
    r = re.sub(r'\W+', ' ', text)
    return r.strip()
```

```python
X = list(map(remove_emoji,list(kb_df['commentText'])))
X = list(map(remove_spechar,X))
y=list(kb_df['Labels'])
```

```python
import torch
from transformers import XLMTokenizer, XLMModel
tokenizer = XLMTokenizer.from_pretrained('xlm-mlm-en-2048')
```

```python
print(X)
```

```python
# Load pre-trained model (weights)
model = XLMModel.from_pretrained('xlm-mlm-en-2048',
                                  output_hidden_states = True # Whether the model returns all hidden-states.
                                  )


# Put the model in "evaluation" mode, meaning feed-forward operation.
model.eval()
                                  
def xlm_vectorization(text):
    # Split the sentence into tokens.    
    inputs = tokenizer(text, return_tensors="pt")
    with torch.no_grad():
        outputs = model(**inputs)
        hidden_states = outputs[1]
    token_vecs = hidden_states[-1][0]
    
    # Calculate the average of all the token vectors.
    sentence_embedding = torch.mean(token_vecs, dim=0)
    
    return np.array(sentence_embedding)
```

```python
xlm_vectors = list(map(xlm_vectorization,X))
```

```python
print(xlm_vectors)
```

```python
xlm_df = pd.DataFrame(xlm_vectors)
xlm_df
```

```python
xlm_df.to_csv(pwd+"//Datasets//Nisha//FineTunedTransformers//xlm_base_finetuned_vectorized_Nisha_dataset.csv",index=False)
```
