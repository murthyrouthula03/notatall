```python
import pandas as pd
import numpy as np
import os,sys
import re
pwd = os.getcwd()
```

```python
# reading dataset
kb_df = pd.read_csv(pwd+"//kabitakitchen.csv", encoding_errors='ignore')
kb_df
```

```python
kb_df.shape
```

```python
kb_df['Labels'] = kb_df['Labels']-1
```

```python
X=list(kb_df['commentText'])
y=list(kb_df['Labels'])
```

```python
print(y)
```

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size = 0.20, random_state = 30)
```

```python
from transformers import GPT2Tokenizer, GPT2Config
# instantiate the configuration for your model
configuration = GPT2Config()
tokenizer = GPT2Tokenizer.from_pretrained('gpt2')
tokenizer.pad_token = tokenizer.eos_token
```

```python
"""if tokenizer.pad_token is None:
    tokenizer.add_special_tokens({'pad_token': '[PAD]'})"""
train_encodings = tokenizer(X_train, truncation=True, padding=True)
test_encodings = tokenizer(X_test, truncation=True, padding=True)
```

```python
print(test_encodings)
```

```python
len(test_encodings['input_ids'][1])
```

```python
import tensorflow as tf

train_dataset = tf.data.Dataset.from_tensor_slices((
    dict(train_encodings),
    y_train
))

test_dataset = tf.data.Dataset.from_tensor_slices((
    dict(test_encodings),
    y_test
))
```

```python
print(test_dataset)
```

```python
from transformers import TFGPT2ForSequenceClassification, TFTrainer, TFTrainingArguments

training_args = TFTrainingArguments(
    output_dir='./GPTresults',          # output directory
    num_train_epochs=2,              # total number of training epochs
    per_device_train_batch_size=8,  # batch size per device during training
    per_device_eval_batch_size=16,   # batch size for evaluation
    warmup_steps=500,                # number of warmup steps for learning rate scheduler
    weight_decay=0.01,               # strength of weight decay
    logging_dir='./GPTlogs',           # directory for storing logs
    logging_steps=10,
    eval_steps=10
)
```

```python
with training_args.strategy.scope():
    model = TFGPT2ForSequenceClassification.from_pretrained("gpt2", num_labels=7)
    # set the pad token of the model's configuration
    model.config.pad_token_id = model.config.eos_token_id

trainer = TFTrainer(
    model=model,                         # the instantiated ðŸ¤— Transformers model to be trained
    args=training_args,                  # training arguments, defined above
    train_dataset=train_dataset,         # training dataset
    eval_dataset=test_dataset             # evaluation dataset
)

trainer.train()
```

```python
trainer.evaluate(test_dataset)
```

```python
output=trainer.predict(test_dataset)
```
